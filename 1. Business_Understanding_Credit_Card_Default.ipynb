# Phân tích và Dự báo Xu hướng Rủi ro Tín dụng Thẻ khách hàng

## 1. Giới thiệu bài toán
Hệ thống tài chính - ngân hàng từ lâu đã được xem là mạch máu của nền kinh tế vĩ mô, phản ánh sức khỏe tài chính nội tại, mức độ tiêu dùng và xu hướng luân chuyển dòng vốn của xã hội. Trong bối cảnh nền kinh tế số phát triển vượt bậc, việc quản lý và phân tích rủi ro tín dụng không còn giới hạn ở các phương pháp chấm điểm truyền thống mà đã chuyển dịch mạnh mẽ sang hướng tiếp cận định lượng bằng khoa học dữ liệu và trí tuệ nhân tạo.

Thẻ tín dụng, với vai trò là công cụ kích cầu tiêu dùng cá nhân phổ biến nhất, sở hữu một quy mô giao dịch và số lượng người dùng khổng lồ, nhưng đồng thời cũng mang đặc tính biến động và rủi ro nợ xấu rất phức tạp. Khả năng thanh toán của khách hàng chịu sự chi phối đan xen của nhiều nhóm yếu tố, từ đặc điểm nhân khẩu học (giới tính, học vấn, hôn nhân, độ tuổi), cho đến các hành vi tài chính thực tế như lịch sử thanh toán trễ hạn, số dư nợ hóa đơn và số tiền đã chi trả trong các tháng liên tiếp. Sự gia tăng đột biến của nợ xấu (Default) tạo ra những rủi ro hệ thống đáng kể cho các tổ chức tài chính, có thể dẫn đến mất cân bằng thanh khoản và suy giảm biên lợi nhuận của ngân hàng.

## 2. Mục tiêu phân tích
Dự án được thực hiện nhằm khai thác bộ dữ liệu quản lý tín dụng thực tế của khách hàng (giai đoạn 5 năm gần nhất, quy mô lớn hơn 30,000 dòng). Mục tiêu cốt lõi là kiến tạo một lăng kính định lượng, giúp các tổ chức tài chính, bộ phận quản trị rủi ro và các nhà hoạch định chiến lược trả lời 3 câu hỏi kinh doanh chiến lược (Research Questions - RQs):

* **RQ1 (Demographic Profiles & Credit Risk):** Rủi ro tín dụng đang phân hóa như thế nào giữa các nhóm khách hàng dựa trên đặc điểm nhân khẩu học? Đâu là những nhóm đối tượng có tỷ lệ trễ hạn cao nhất? Yếu tố độ tuổi và trình độ học vấn ảnh hưởng ra sao đến khả năng thực hiện nghĩa vụ tài chính?
* **RQ2 (Behavioral Dynamics & Payment History):** Dựa trên lịch sử hành vi thanh toán qua các tháng, các dấu hiệu trễ hạn sớm (như nợ quá hạn từ 1 đến 2 tháng) có mối liên hệ dịch chuyển như thế nào đến khả năng quỵt nợ hoàn toàn trong tương lai? Phân khúc hạn mức tín dụng (Credit Limit) nào đang có rủi ro cao nhất?
* **RQ3 (Financial Features & Classification Efficiency):** Các yếu tố tài chính định lượng (số tiền hóa đơn phát sinh, tỷ lệ thanh toán tối thiểu, số dư nợ lũy kế) đóng góp như thế nào vào mô hình đánh giá rủi ro tín dụng? Có thể xây dựng mô hình phân loại học máy chính xác khả năng vỡ nợ (Default vs Non-Default) dựa trên các đặc trưng này hay không?

Kết quả phân tích phục vụ cho quá trình ra quyết định dựa trên dữ liệu (Data-driven decision making) của 3 nhóm đối tượng chính:
* **Bộ phận Phê duyệt & Quản trị Rủi ro Tín dụng tại Ngân hàng:** Xác định các tệp khách hàng có nguy cơ nợ xấu cao để chủ động thắt chặt hạn mức, tối ưu hóa danh mục cho vay an toàn.
* **Bộ phận Phát triển Sản phẩm & Kinh doanh Tài chính:** Xây dựng chính sách ưu đãi, chấm điểm tín dụng tự động để hỗ trợ phê duyệt hồ sơ mở thẻ nhanh chóng, minh bạch cho khách hàng tốt.
* **Cơ quan Quản lý & Giám sát Ngân hàng Nhà nước:** Nắm bắt xu hướng tiêu dùng và nợ xấu để đưa ra các chính sách quản lý trần tín dụng phù hợp, đảm bảo an toàn hệ thống tài chính quốc gia.

## 3. Các yếu tố định hình rủi ro tín dụng
Dựa trên bản chất của bộ dữ liệu quản lý tín dụng thực tế (quy mô 30,000 bản ghi), có các yếu tố trọng yếu trực tiếp chi phối và định hình rủi ro nợ xấu của khách hàng:
* **Hạn mức tín dụng được cấp (LIMIT_BAL):** Phản ánh tiềm lực tài chính ban đầu hoặc độ uy tín của khách hàng mà ngân hàng đánh giá.
* **Đặc điểm nhân khẩu học (SEX, EDUCATION, MARRIAGE, AGE):** Phản ánh các yếu tố nền tảng về giới tính, trình độ học vấn, tình trạng hôn nhân và độ tuổi — những biến số xã hội học thường có sự tương quan lớn đối với ý thức và hành vi quản lý dòng tiền cá nhân.
* **Lịch sử trạng thái thanh toán (PAY_0 đến PAY_6):** Các chỉ số theo dõi trạng thái trả nợ hàng tháng (trả đúng hạn, trễ 1 tháng, trễ 2 tháng,...). Đây là nhóm thuộc tính động quan trọng nhất để nhận diện rủi ro sớm.
* **Số dư hóa đơn phát sinh (BILL_AMT1 đến BILL_AMT6):** Số dư nợ thẻ phát sinh qua từng kỳ sao kê, phản ánh mức độ chi tiêu và gánh nặng nợ thực tế của khách hàng.
* **Số tiền đã thanh toán thực tế (PAY_AMT1 đến PAY_AMT6):** Số tiền khách hàng đã chủ động nộp lại cho ngân hàng trong các kỳ trước để giảm dư nợ, thể hiện tính thanh khoản dòng tiền cá nhân.

## 4. Cách tiếp cận phân tích
Để giải quyết triệt để 3 câu hỏi cốt lõi, dự án áp dụng quy trình phân tích dữ liệu chuẩn mực và khắt khe qua các bước:
1. **Làm sạch dữ liệu (Data Cleaning):** Kiểm tra và xử lý các giá trị khuyết khuyết, loại bỏ dữ liệu nhiễu, mã hóa biến phân loại và đồng bộ các giá trị không nhất quán trong hệ thống.
2. **Phân tích thành phần chính (PCA) & Trích xuất đặc trưng:** Sử dụng kỹ thuật giảm chiều dữ liệu PCA trên các nhóm biến có độ tương quan cao (như các cột Bill Amount và Pay Amount) nhằm tối ưu không gian đặc trưng và hạn chế hiện tượng quá khớp (Overfitting).
3. **Phân chia dữ liệu trong học máy (Data Splitting):** Chia tập dữ liệu 30,000 dòng theo tỷ lệ 80/20 (Train/Test) kết hợp kỹ thuật lấy mẫu SMOTE để xử lý hiện tượng mất cân bằng dữ liệu (Imbalanced Data) giữa nhóm khách hàng nợ xấu và khách hàng tốt.
4. **Lựa chọn mô hình học máy (Model Selection):** Thử nghiệm và đối chứng các thuật toán phân loại đa dạng bao gồm Logistic Regression (Baseline), Random Forest (Tree-based) và đặc biệt là các thuật toán Boosting mạnh mẽ như XGBoost và LightGBM.
5. **Huấn luyện mô hình (Model Training):** Tiến hành tối ưu hóa mô hình qua kỹ thuật dò tìm siêu tham số Cross-Validation và Grid Search nhằm tìm ra bộ tham số tốt nhất cho thuật toán dự báo rủi ro.
6. **Đánh giá mô hình học máy (Model Evaluation):** Đánh giá hiệu năng dựa trên các chỉ số nâng cao như Precision, Recall (Độ nhạy bắt nợ xấu), F1-Score và đường cong AUC-ROC để chứng minh tính vượt trội của giải pháp đề xuất.
7. **Triển khai mô hình học máy (Model Deployment):** Đóng gói mô hình tối ưu nhất thành file định dạng `.pkl` và thiết lập kiến trúc API giả định để tích hợp trực tiếp vào hệ thống xét duyệt tín dụng tự động theo thời gian thực (Real-time).

## 5. Mục tiêu kỳ vọng
* Cung cấp minh chứng định lượng cụ thể về sự phân hóa rủi ro tín dụng dựa trên hành vi và nhân khẩu học của khách hàng, đặc biệt là mối liên hệ giữa các dấu hiệu trễ hạn sớm với khả năng vỡ nợ thực tế.
* Nhận diện chính xác các nhóm thuộc tính chi tiêu và lịch sử hóa đơn có tương quan mạnh tới nợ xấu, giúp ngân hàng định hướng chiến lược quản trị danh mục cho vay an toàn.
* Xây dựng thành công mô hình phân loại học máy (XGBoost, LightGBM) dự đoán nợ xấu với độ chính xác cao và chỉ số Recall tối ưu (trên 75%), hỗ trợ phát hiện rủi ro một cách tự động.
* Thiết lập một Khung phân tích dữ liệu rủi ro tài chính thực chiến (Data-driven Credit Risk Framework) làm cơ sở cho các quyết định duyệt cấp thẻ và kiểm soát nợ xấu của ngân hàng trong tương lai.
