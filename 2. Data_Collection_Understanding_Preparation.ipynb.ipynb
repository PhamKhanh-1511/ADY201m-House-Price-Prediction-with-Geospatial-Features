```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

# 1. Đọc dữ liệu từ file local của dự án
print("Đang tải dữ liệu...")
df = pd.read_csv('UCI_Credit_Card.csv')

# 2. Làm sạch khoảng trắng dư thừa ở tên các cột
df.columns = df.columns.str.strip()

# 3. Tạo ánh xạ và chuyển đổi nhãn danh mục (Tương tự như map Quarter ở ví dụ)
# Ánh xạ trình độ học vấn từ số sang dạng chuỗi text tường minh
education_map = {
    1: 'Graduate_School',
    2: 'University',
    3: 'High_School',
    4: 'Others',
    0: 'Others',
    5: 'Others',
    6: 'Others'
}
df['Education_Level'] = df['EDUCATION'].map(education_map)

# 4. Định nghĩa và ép kiểu các cột số nguyên (int64)
int_cols = [
    'LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE',
    'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6',
    'default.payment.next.month'
]

for col in int_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype('int64')

# 5. Định nghĩa và ép kiểu các cột số thực (float64)
float_cols = [
    'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6',
    'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6'
]

for col in float_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0.0)

# 6. Tính toán tạo tính năng mới: Tỷ lệ thanh toán trên hóa đơn kỳ gần nhất (Giống Logistics_Cost_Ratio)
if 'Pay_To_Bill_Ratio' not in df.columns and 'PAY_AMT1' in df.columns:
    df['Pay_To_Bill_Ratio'] = df['PAY_AMT1'] / df['BILL_AMT1']
    # Xử lý các giá trị vô cùng (chia cho 0) hoặc giá trị trống rỗng rác phát sinh
    df['Pay_To_Bill_Ratio'] = df['Pay_To_Bill_Ratio'].replace([np.inf, -np.inf], 0).fillna(0.0)

# 7. Hiển thị thông tin kiểm tra kiểu dữ liệu
print("\n--- KIỂU DỮ LIỆU CÁC CỘT ---")
print(df.dtypes)

# 8. Thống kê mô tả tổng quan (Loại bỏ cột định danh ID giống như bỏ SITC Number)
print("\n--- THỐNG KÊ MÔ TẢ (Đã loại bỏ cột mã ID) ---")
stats_overview = df.describe()
stats_overview = stats_overview.drop('ID', axis=1, errors='ignore')
display(stats_overview.T)

# 9. Đồng bộ định dạng tên cột: Thay thế dấu chấm và dấu cách bằng dấu gạch dưới (_)
df.columns = df.columns.str.replace('.', '_').str.replace(' ', '_')

# 10. Lưu kết quả làm sạch ra file mới theo luồng xử lý
output_file = 'credit_card_prepared.csv'
df.to_csv(output_file, index=False)
print(f"\n[THÀNH CÔNG] Dữ liệu đã được làm sạch và lưu thành file: '{output_file}'")
```
